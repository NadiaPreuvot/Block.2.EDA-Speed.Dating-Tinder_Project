# 2.<u>Data cleaning

1. <u> Replace the numerical values by their actual names

```python
#Replacing the gender indexes by their actual names
sex = {0 : 'female', 1: 'male'}
df['gender'] = df['gender'].map(sex)
```

```python
# Replacing Match: 1 by yes and 0 by no 
df["match"]=df["match"].replace({1: "yes" , 0: "no"})
```

```python
#Replacing the value for participant and the partner by 1 = yes if they were the same race or else by 0=no
df["samerace"]= df["samerace"].replace({1 : "yes", 0: "no"})
```

2. <u>Let's change the numbers into names to better understand the data.

```python
#replace the numerical value by their inial name for the race column
#1.creating a dictionnary to replace
race_list = { 
    1: "Black/African American", 
    2: "European/Caucasian-American",
    3: "Latino/Hispanic American", 
    4: "Asian/Pacific Islander/Asian-American",
    5: "Native American", 
    6 : "Other"  
}

#2. Replace the numbers by their acutal names
df["race"]=df["race"].map(race_list)
```

```python
# Replace the values(numbers) of career into their actual names
#Creating a dictionnary to replace
df['field_cd'] = df['field_cd'].map({
    1: "Law",
    2: " Math", 
    3: "Social Science, Psychologist " ,
    4: "Medical Science, Pharmaceuticals, and Bio Tech" ,
    5: "Engineering " ,
    6: "English/Creative Writing/ Journalism ",
    7: "History/Religion/Philosophy",
    8:  "Business/Econ/Finance ",
    9:  "Education, Academia" ,
    10: "Biological Sciences/Chemistry/Physics",
    11: "Social Work",
    12: "Undergrad/undecided",
    13: "Political Science/International Affairs ",
    14: "Film",
    15: "Fine Arts/Arts Administration",
    16: "Languages",
    17: "Architecture",
    18: "Other"
})
```

```python
#Replace the attributes diminutive by their full names.

attributes_rename = {
    "attr1_1": "Attractive",
    "sinc1_1": "Sincere",
    "intel1_1": "Intelligent",
    "fun1_1": "Fun",
    "amb1_1": "Ambitious",
    "shar1_1": "shared interests"
}
#Mapping 
df.rename(columns= attributes_rename , inplace=True)
```

8. <u> How Does Race Affect Match Success in Speed Dating? / What is the match rate based on race ?

```python
# Color for each race
colors = ['blue', 'orange', 'green', 'red', 'purple']  
# Mapping  
df['match'] = df['match'].map({'yes': 1, 'no': 0}).astype(int)

# Get Data and sorting for each race
race_counts = df['race'].value_counts().sort_index()

# Calculer les taux moyens de match par race et les ordonner de la même manière
race_match_rates = df.groupby('race')['match'].mean().sort_index()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 8))

# first graph : Repartition of race
axes[0].pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
axes[0].set_title('Repartition of the different race', fontweight='bold')
axes[0].set_ylabel('')  # drop axe Y

# Second graph : match per race
axes[1].bar(race_match_rates.index, race_match_rates, color=colors)
axes[1].set_title('Match Rate by Race', fontweight='bold')
axes[1].set_xlabel('Race')
axes[1].set_ylabel('Match Rate')
axes[1].set_ylim(0, 0.3)  
axes[1].set_xticks(range(len(race_match_rates.index)))  # align the tags
axes[1].set_xticklabels(race_match_rates.index, rotation=45, ha='right')  # Rotate

plt.tight_layout()
plt.show()
```

</u><ins><center> <font color="bleu"><ins>Observations</font> </center>

These visualizations highlight an interesting aspect of racial dynamics in speed dating outcomes. Despite European/Caucasian-Americans constituting the majority of participants, as shown in the pie chart, they do not have the highest match rates, which are depicted in the bar chart. This discrepancy suggests that being in the racial majority does not confer a matching advantage in the speed dating context. <br>
It implies that participants might prioritize other qualities over racial similarity when choosing potential dates. <br>This observation can challenge assumptions about racial preferences in dating and suggests that individual attributes and interpersonal chemistry likely play a more significant role in match decisions. <br>

9 . <u> Is there a correlation between participants' age and their match decisions ?

```python
# Correlation between participants' age and match decisions

# Select columns and drop missing values
df_age = df[['match', 'age']].dropna().copy()

# Creat boxplot match based on age
plt.figure(figsize=(18, 6))
sns.boxplot(x='match', y='age', data=df_age)
plt.title('Participant Age Based on Match Decisions', weight='bold')
plt.xlabel('Match Decision (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.tight_layout()
plt.show()

# Calculating the Spearman's correlation coefficient
correlation = df_age['age'].corr(df_age['match'], method='spearman')
print(f"Spearman's correlation coefficient between 'age' and 'match' is: {correlation:.2f}")
```

</u><ins><center> <font color="bleu"><ins>Observations</font> </center>

This boxplot illustrates the distribution of participants' ages based on their match decisions during speed dating events. The median age is similar across both groups who did and did not form a match, suggesting that age alone may not be a significant factor in determining match outcomes. The presence of outliers, particularly in the 'No Match' group, indicates greater age variability among those who did not find a match. The Spearman's correlation coefficient between 'age' and 'match' is -0.03, reinforcing that there is a very weak negative correlation between age and matching success. These findings imply that factors other than age might be more influential in determining the success of matches at these events.

10.<u> Does the age difference between partners influence the match rate?

```python
# Selecting the columns 'match', 'age', and 'age_o' et suppression des valeurs manquantes
df_age_diff = df[['match', 'age', 'age_o']].dropna().copy()

# Count age difference  and stock in a new column
df_age_diff['diff_age'] = (df_age_diff['age'] - df_age_diff['age_o']).abs()

## Plotting a box plot to visualize the distribution of age difference based on match
plt.figure(figsize=(18, 6))
sns.boxplot(x="match", y="diff_age", data=df_age_diff)
plt.title("Distribution of Absolute Age Difference vs. Match", weight = "bold")
plt.xlabel("Match Decision")
plt.ylabel("Absolute Age Difference")
plt.tight_layout()
plt.show()

# Calculating the Spearman's correlation coefficient entre la différence d'âge absolue et la décision de match
correlation = df_age_diff['diff_age'].corr(df_age_diff['match'], method='spearman')
print(f"Spearman's correlation coefficient between absolute age difference and match is: {correlation:.2f}")
```

</u><ins><center> <font color="bleu"><ins>Observations</font> </center>

The box plot shows the distribution of absolute age differences between partners in speed dating, with a focus on match outcomes.<br> Both groups—those who matched and those who didn't—exhibit similarly low median age differences, indicating that age disparity has minimal impact on matching success.<br> The Spearman correlation coefficient of -0.07 confirms this, revealing a very weak negative correlation between age difference and match likelihood.<br> This suggests that age difference is not a significant factor in determining match outcomes in speed dating.

17. <u> Is there any benefit to communicating with a partner where a match occurred following the event ?

```python
# Selecting and renaming columns
df_call = df[['you_call', 'them_cal', 'match', 'date_3']]

# Filtering rows where 'match_numeric' is equal to 1
df_callmatch = df_call[df_call['match'] == 1].copy()

# Added 'call_made' column and removed missing values
df_callmatch['call_made'] = df_callmatch['you_call'] + df_callmatch['them_cal']
df_callmatch_clean = df_callmatch.dropna()

# Calculation of the Spearman correlation between 'call_made' and 'date_3'
spearman_correlation = df_callmatch_clean['call_made'].corr(df_callmatch_clean['date_3'], method='spearman')

# Creating a regplot with a regression line
plt.figure(figsize=(18, 6))
sns.regplot(x='call_made', y='date_3', data=df_callmatch_clean, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line between Calls Made and Going on a Date')
plt.xlabel('Total Calls Made (You + Them)')
plt.ylabel('Went on a Date (date_3)')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.tight_layout()
plt.show()

# Displaying Spearman correlation
print(f"Spearman's Correlation between 'call_made' and 'date_3': {spearman_correlation}")
```

</u><ins><center> <font color="bleu"><ins>Observations</font> </center>
</br>
The scatter plot illustrated above showcases the relationship between the total number of calls made (sum of calls initiated by the participant and received from their match) and the likelihood of going on a second date. <br>
The red line represents a regression line that predicts the likelihood of securing a second date based on the volume of communication. 
Each blue dot represents a specific case, positioned by the total calls made on the X-axis and the outcome of going on a second date on the Y-axis, where 1 indicates a date occurred and 0 indicates none.

 The regression line suggests a positive correlation, indicates that as the number of calls increases, so does the probability of participants deciding to meet again.<br>
This is further supported by a Spearman's correlation coefficient of 0.536, demonstrating a strong positive correlation between post-event communication and the intention to pursue a second date. 

This observation underscores the significant role that communication plays in turning matches into actual dates, highlighting the importance of follow-up interactions in the dating process.
