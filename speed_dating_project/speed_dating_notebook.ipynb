# 5. <u> Multivariate Analysis

1. <u >What attributes are most strongly correlated with getting a 'yes' for a second date as per the partner's opinion ?

```python
# Creating a new DataFrame with the relevant columns
df_dec = df[['gender', 'dec_o', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']].copy()

# Converting 'gender' to numeric values ​​with a lambda function: 'female' becomes 0, 'male' becomes 1
df_dec['gender'] = df_dec['gender'].apply(lambda x: 0 if x == 'female' else 1)

# Defining attribute columns
attributes_columns = ['attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']

# Eliminating rows with missing values
df_dec = df_dec.dropna()

# Calculation of the correlation between each attribute and 'dec_o' for each genre
correlations_attributes_dec_o = {}
for column in attributes_columns:
    correlations_attributes_dec_o[column] = (
        df_dec[df_dec['gender'] == 0][[column, 'dec_o']].corr().iloc[0, 1],
        df_dec[df_dec['gender'] == 1][[column, 'dec_o']].corr().iloc[0, 1]
    )

# Converting Correlation Dictionaries to DataFrame
correlations_df = pd.DataFrame(correlations_attributes_dec_o, index=['Women', 'Men']).T

new_colors = ['#ff69b4', '#6495ed']

# Creating a correlation bar chart
correlations_df.plot(kind='bar', figsize=(15, 10), alpha=0.7, color=new_colors)
plt.title('Correlation between score given by partner and their decision for a second date',  weight='bold')
plt.xlabel('Attribute rated by partner')
plt.ylabel('Rate of yes (dec_o = 1)')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

</u><ins><center> <font color="bleu"><ins>Observations</font> </center>
</br>
The analysis reveals that for both genders, the traits of physical appeal (attr_o), enjoyment (fun_o), and common interests (shar_o) play significant roles in achieving a second date. Particularly, the physical appeal factor exhibits a very strong correlation (over 50%) with the likelihood of women securing a second date.

3. <u> Is there any benefit to communicating with a partner where a match occurred following the event ?

```python
# Conversion de 'match' en valeur numérique
df['match_numeric'] = df['match'].apply(lambda x: 1 if x == 'yes' else 0)

# Selecting and renaming columns
df_call = df[['you_call', 'them_cal', 'match_numeric', 'date_3']]

# Filtering rows where 'match_numeric' is equal to 1
df_callmatch = df_call[df_call['match_numeric'] == 1].copy()

# Added 'call_made' column and removed missing values
df_callmatch['call_made'] = df_callmatch['you_call'] + df_callmatch['them_cal']
df_callmatch_clean = df_callmatch.dropna()

# Calculation of the Spearman correlation between 'call_made' and 'date_3'
spearman_correlation = df_callmatch_clean['call_made'].corr(df_callmatch_clean['date_3'], method='spearman')

# Creating a regplot with a regression line
plt.figure(figsize=(20, 10))
sns.regplot(x='call_made', y='date_3', data=df_callmatch_clean, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line between Calls Made and Going on a Date')
plt.xlabel('Total Calls Made (You + Them)')
plt.ylabel('Went on a Date (date_3)')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.tight_layout()
plt.show()

# Displaying Spearman correlation
print(f"Spearman's Correlation between 'call_made' and 'date_3': {spearman_correlation}")
```

</u><ins><center> <font color="bleu"><ins>Observations</font> </center>
</br>
The scatter plot illustrated above showcases the relationship between the total number of calls made (sum of calls initiated by the participant and received from their match) post-speed dating event and the participants' decision to go on a second date. The red line represents a regression line that predicts the likelihood of securing a second date based on the volume of communication. 

A positive slope of this line indicates that as the number of calls increases, so does the probability of participants deciding to meet again. This is further supported by a Spearman's correlation coefficient of 0.5361475174558188, demonstrating a strong positive correlation between post-event communication and the intention to pursue a second date. 

This observation underscores the significant role that communication plays in turning matches into actual dates, highlighting the importance of follow-up interactions in the dating process.

5. <u> Is there a correlation between the attractiveness ratings participants give themselves and the ratings given by partners ?
